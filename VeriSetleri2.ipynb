```python
veri=pd.read_csv("eksikveriler.csv")
from sklearn.impute import SimpleImputer
eksikveriTamamlama=SimpleImputer(missing_values=np.nan, strategy='mean')
eksikveriTamamlama.fit(veri.iloc[:,1:4])
veri.iloc[:,1:4]=eksikveriTamamlama.transform(veri.iloc[:,1:4])
veri
```
